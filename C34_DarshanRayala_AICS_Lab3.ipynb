```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/50_Startups.csv')

# Spliting Dataset in Dependent & Independent Variables
X = df.iloc[:, :-1].values
y= df.iloc[:, 4].values

labelencoder=LabelEncoder()
X[:,3]=labelencoder.fit_transform(X[:,3])
x_train,x_test,y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=0)
model= LinearRegression()
model.fit(x_train, y_train)

print("bo:", model.intercept_)

print("Coefficients:", model.coef_)

y_pred=model.predict(x_test)

testing_data_model_score = model.score (x_test, y_test) 
print("Model Performance on Testing data: ", testing_data_model_score)

training_data_model_score = model.score (x_train, y_train) 
print("Model Performance on Training data: ", training_data_model_score)

df1 = pd.DataFrame(data ={'Predicted value':y_pred.flatten(), 'Actual Value':y_test. flatten()})
print("Prediction using the model:")
df1.head(10)
```
